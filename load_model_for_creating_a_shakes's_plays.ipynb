# **Loading the model (poetry_creation_h5) for the prediction/generating a poem**

```
from tensorflow import keras

model_path = "/content/poetry_creation_model.h5"

from keras.models import load_model

model = load_model(model_path)
```

```
model.summary()
```

# **but before prediction we need to download the plays and tokenize it first**

```
# links

!wget --no-check-certificate \
    https://storage.googleapis.com/laurencemoroney-blog.appspot.com/sonnets.txt \
    -O /tmp/sonnets.txt
data = open('/tmp/sonnets.txt').read()
```

```
from tensorflow import keras
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```
datas = data.lower().split("\n")
tokenizer = Tokenizer()
tokenizer.fit_on_texts(datas)
word_index = tokenizer.word_index
```

```
word_index
```

```
# max_len_sequence

max_len_sequence = 11

def prediction_processing(model, text, no_words):
  for i in range(no_words):
    token_list = tokenizer.texts_to_sequences([text])[0]
  
    token_list = pad_sequences([token_list], maxlen= max_len_sequence-1, padding= "pre")
    predicted = model.predict_classes(token_list, verbose = 0)
    output = ""
    for words, index in tokenizer.word_index.items():
      if index == predicted:
        output = words
        break
    text +=" "+ output
# combine the result with output words

  print(text)


text = "i were a lazy one"

# how many number of words you want to generate

no_word = 20
prediction_processing(model, text, no_word)


```
