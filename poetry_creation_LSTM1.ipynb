```
import tensorflow
from tensorflow import keras
from keras.preprocessing.text import Tokenizer
```

```
tokenizer = Tokenizer()
data
```

```
datas = data.lower()
datas = datas.split("\n")
```

```
datas
```

```
from keras.preprocessing.sequence import pad_sequences

tokenizer.fit_on_texts(datas)
word_index = tokenizer.word_index
```

```
word_index
```

```
total_words = len(word_index)+1
total_words
```

```
input_sequence = []

for i in datas:
  token_list = tokenizer.texts_to_sequences([i])[0]
  print(token_list)
  for i in range(1,len(token_list)):
    print(i)
    n_gram_seq = token_list[:i+1]
    print(n_gram_seq)
  break
```

```
input_sequence = []

for i in datas:
  token_list = tokenizer.texts_to_sequences([i])[0]

  for i in range(1,len(token_list)):
    
    n_gram_seq = token_list[:i+1]
    input_sequence.append(n_gram_seq)
```

```
input_sequence
```

```
# getting the maximum valuer from the passage

max_seq_len = max([len(x) for x in input_sequence])
max_seq_len
```

```
# sequence padding
import numpy as np
input_sequences = np.array(pad_sequences(input_sequence, maxlen = max_seq_len, padding = "pre"))
input_sequences
```

```
# splitiing the x and y labels
X = input_sequences[:,:-1]
print(X)
y = input_sequences[:,-1]
print(y)
```

```
# label for y

y = keras.utils.to_categorical(y)
print(y)
len(y)
```

```
# model.define

model = keras.Sequential([
keras.layers.Embedding(total_words, 64, input_length= max_seq_len-1),
keras.layers.Bidirectional(keras.layers.LSTM(20)),
keras.layers.Dense(total_words, activation= "softmax")                          
        ])
```

```
model.compile(
    optimizer = "adam",
    loss = "categorical_crossentropy",
    metrics = ["accuracy"]
)
```

```
history = model.fit(X,y , epochs = 100, verbose = 2)
```

```
# prediction words results

text_to_pred = "jenifier you're my only hope"
next_word = 20

for i in range(next_word):
  token_list = tokenizer.texts_to_sequences([text_to_pred])[0]
  print(token_list)

  token_lists = pad_sequences([token_list], maxlen= max_seq_len-1, padding="pre")
  print(token_lists)
```

```
text_to_pred = "jenifier where she is alone"
next_word = 100

for i in range(next_word):
  token_list = tokenizer.texts_to_sequences([text_to_pred])[0]
  
  token_lists = pad_sequences([token_list], maxlen= max_seq_len-1, padding="pre")
  predicted = model.predict_classes(token_lists, verbose = 0)
  
  output = ""
  for words, index in tokenizer.word_index.items():
    if index == predicted:
      output = words
      break
  text_to_pred +=" "+ output
# combine the result with output words

print(text_to_pred)

```
